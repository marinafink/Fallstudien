```python
# Import Tweets frpm Joe Biden
df = pd.read_csv("tweets.csv")

# Show first 3 rows
df.head(3)
```

```python
# Transform strings to lower case
df['text'] = df['text'].astype(str).str.lower()

df.head(3)
```

```python
from nltk.tokenize import RegexpTokenizer

regexp = RegexpTokenizer('\w+')

# Create a new column called text_token from text column
# apply the function regexp
df['text_token']=df['text'].apply(regexp.tokenize)

df.head(3)
```

```python
import nltk

nltk.download('stopwords')
```

```python
import nltk
from nltk.corpus import stopwords

# Make a list of english stopwords
stopwords = nltk.corpus.stopwords.words("english")

# Extend the list with your own custom stopwords
# add the word https
my_stopwords = ['https']

stopwords.extend(my_stopwords)
```

```python
stopwords
```

```python
df['text_token'] = df['text_token'].apply(lambda x: [item for item in x if item not in stopwords])

df.head(3)
```

```python
df['text_token'] = df['text_token'].apply(lambda x: ' '.join([item for item in x if len(item)>2]))

df.head(3)
```

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
#nimm aus dem Submodul stem aus der library nltk die Funktion WordNetLemm

wordnet_lem = WordNetLemmatizer()

# use on text_token
df['text_token'] = df['text_token'].apply(wordnet_lem.lemmatize)
```

```python
df
```

```python
all_words = ' '.join([word for word in df['text_token']])
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
```

```python
# Create column called polarity
df['polarity'] = df['text_token'].apply(lambda x: analyzer.polarity_scores(x)) #neue Spalte polarity im dataframe erzeugen

df.head(3)
```

```python
# Change data structure
df = pd.concat(
    [df.drop(['Unnamed: 0', 'id', 'author_id', 'polarity'], axis=1), 
     df['polarity'].apply(pd.Series)], axis=1)

df.head(3)
```

```python
# Create new variable with sentiment "neutral," "positive" and "negative"
# Call the new column sentiment 
df['sentiment'] = df['compound'].apply(lambda x: 'positive' if x >0 else 'neutral' if x==0 else 'negative')

df.head(4)
```

```python
# Tweet with highest positive sentiment (use max)
df.loc[df['compound'].idxmax()].values
```

```python
# Tweet with highest negative sentiment (use min)
# ...seems to be a case of wrong classification because of the word "deficit"
df.loc[df['compound'].idxmin()].values
```

```python
# Number of tweets with certain sentiment
sns.countplot(y='sentiment', 
             data=df, 
             palette=['#b2d8d8',"#008080", '#db3d13']
             );
```

```python
# Lineplot with compound
g = sns.lineplot(x='created_at', y='sentiment', data=df)

g.set(xticklabels=[]) #nimmt die labels unten an der Achse raus
g.set(title='Sentiment of Tweets')
g.set(xlabel="Time")
g.set(ylabel="Sentiment")
g.tick_params(bottom=False)

g.axhline(0, ls='--', c = 'grey');
```

```python
# Boxplot
sns.boxplot(y='compound', 
            x='sentiment',
            palette=['#b2d8d8',"#008080", '#db3d13'], 
            data=df);
```
